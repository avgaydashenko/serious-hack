```python
import pandas as pd
import numpy as np
```

```python
FILE_NAME = "test"
df = pd.read_csv("inspection_schedules_{}_hackathon_20jan18.csv.zip".format(FILE_NAME))
```

```python
df
```

```python
df.dtypes
```

```python
df.loc[df.city.notnull(), 'type_nasel_punkt'] = 'Город'
df.loc[df.city.notnull(), 'name_nasel_punkt'] = df[df.city.notnull()].city
for name in ['region', 'company_type', 'company_name', 'okved_osn_name', 'init_equity_types', 'purpose', 'type', 'agency', 'joint_agency', 'type_nasel_punkt', 'name_nasel_punkt', 'street']:
    df[name] = df[name].str.lower()
df = df.drop(columns=['city'])
```

```python
df
```

```python
df.loc[df.type == 'в', 'type'] = 'выездная'
df.loc[df.type == 'д', 'type'] = 'документарная'
df.loc[df.type == 'дв', 'type'] = 'документарная и выездная'
df.loc[df.type == 'документарная, выездная', 'type'] = 'документарная и выездная'
df.loc[df.type == 'выездная докуметарная', 'type'] = 'документарная и выездная'
df.loc[df.type == '0', 'type'] = np.NaN
df.type.unique()
```

```python
df
```

```python
for name in ['ogrn', 'inn', 'okved_osn_code', 'oktmo', 'okato', 'kladr', 'kpp', 'index']:
    df[name] = df[name].astype('str')
df = df.drop(columns=['year'])
```

```python
df.to_csv('{}.csv.gzip'.format(FILE_NAME), compression='gzip')
```
