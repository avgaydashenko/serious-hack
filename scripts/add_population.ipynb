```python
import numpy as np
import pandas as pd
%matplotlib inline
import seaborn as sns
from tqdm import tqdm
import matplotlib.pyplot as plt
sns.set(color_codes=True)
```

```python
data_train = pd.read_csv("../train.csv")
data_test = pd.read_csv("../test.csv")
```

# Merge with population

```python
population = pd.read_csv("../population.csv")
```

```python
population.head()
```

Try to extract all citites that are set as, eg,  г.Адыгея.

```python
pop_values = population[["municipalities", "total", "urban", "rural"]]
pop_values["municipalities"] = pop_values.loc[pop_values["municipalities"].apply(lambda x: len(x.split('.')) > 1)]
pop_values.head()
```

```python
pop = pop_values[~pop_values["municipalities"].isnull()]
```

```python
pop.head()
```

```python
pop["municipalities"] = pop["municipalities"].apply(lambda x: x.split('.')[1].strip().lower())
```

```python
pop = pop.groupby("municipalities").sum()
pop.head()
```

```python
list(pop.index)
```

```python
data_train_join = data_train.join(pop, on=["name_nasel_punkt"])
data_train_join
```

```python
data_test_join = data_test.join(pop, on=["name_nasel_punkt"])
data_test_join
```

```python
print(data_train_join.shape[0])
print(data_train_join["total"].isnull().sum())
```

```python
print(data_test_join.shape[0])
print(data_test_join["total"].isnull().sum())
```

```python
def join_rest(data_join):
    inds_total_null = data_join[data_join["total"].isnull()].index
    inds_city_not_null = data_join.loc[inds_total_null].loc[~data_join.loc[inds_total_null]["name_nasel_punkt"].isnull()].index
    for ind in tqdm(inds_city_not_null):
        city = data_join.loc[ind]["name_nasel_punkt"]
        rows_for_city = population.loc[population["municipalities"].apply(lambda s: s.lower().find(city) != -1)]
        if rows_for_city.shape[0] != 0:
            values = rows_for_city[["total", "urban", "rural"]].sum()
            data_join.set_value(ind, "total", values["total"])
            data_join.set_value(ind, "urban", values["urban"])
            data_join.set_value(ind, "rural", values["rural"])
```

```python
join_rest(data_train_join)
```

```python
join_rest(data_test_join)
```
