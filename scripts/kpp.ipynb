```python
kpp1 = pd.read_csv('kpp1.csv', sep=';', encoding='1251')
```

```python
kpp2 = pd.read_csv('kpp2.csv', sep=';')
```

```python
kpp3 = pd.read_csv('kpp3.csv', sep=';', encoding='1251')
```

```python
test['kpp'] = test['kpp'] // 1000
train['kpp'] = train['kpp'] // 1000
```

```python
test_code = test['kpp'] // 100
train_code = train['kpp'] // 100
```

```python
test['kpp'] = test['kpp'] % 100
train['kpp'] = train['kpp'] % 100
```

```python
result = []
not_found = []
for v in test_code:
    if isnan(v):
        result.append(np.nan)
    else:
        v = int(v)
        tmp = kpp1['ADDRESS'][kpp1['KOD'] == v].values
        if len(tmp) != 0 and type(tmp[0]) == str:
            result.append(tmp[0][1:7])
        else:
            tmp = kpp2['G1'][kpp2['GA'] == v].values
            if len(tmp) == 0:
                tmp = kpp3['G1'][kpp3['GA'] == v].values
            if len(tmp) == 0:
                not_found.append(int(v))
                result.append(np.nan)
            else:
                result.append(tmp[0][1:7])
```
